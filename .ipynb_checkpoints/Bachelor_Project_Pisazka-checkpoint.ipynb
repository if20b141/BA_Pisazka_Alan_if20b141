```python
import os
import numpy as np
import timm
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from torch.utils.data import TensorDataset, DataLoader
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
import itertools
```

```python
project_path = './'
image_size = (128, 128)  # Passen Sie die Größe an Ihre Anforderungen an
batch_size = 64

datagen = ImageDataGenerator(rescale=1./127)

train_data = datagen.flow_from_directory(
    os.path.join(project_path, 'archive/Training'),
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_data = datagen.flow_from_directory(
    os.path.join(project_path, 'archive/Testing'),
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

```python
lenet5 = keras.Sequential([
    layers.Conv2D(6, (5, 5), activation='relu', input_shape=(128, 128, 3)),
    layers.AveragePooling2D(),
    layers.Conv2D(16, (5, 5), activation='relu'),
    layers.AveragePooling2D(),
    layers.Flatten(),
    layers.Dense(120, activation='relu'),
    layers.Dense(84, activation='relu'),
    layers.Dense(4, activation='softmax')  # Anzahl der Klassen anpassen
])

lenet5.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

alexnet = keras.Sequential([
    layers.Conv2D(96, (11, 11), strides=(4, 4), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2)),
    layers.Conv2D(256, (5, 5), padding='same', activation='relu'),
    layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2)),
    layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
    layers.Conv2D(384, (3, 3), padding='same', activation='relu'),
    layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
    layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2)),
    layers.Flatten(),
    layers.Dense(4096, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(4096, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(4, activation='softmax')  # Anzahl der Klassen anpassen
])

alexnet.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

base_model_inception = InceptionV3(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

# Fügen Sie Ihre eigenen Schichten hinzu
inception_model = keras.Sequential([
    base_model_inception,
    GlobalAveragePooling2D(),
    Dense(512, activation='relu'),
    Dense(4, activation='softmax')  # Anzahl der Klassen anpassen
])

# Kompilieren Sie das Inception-v3-Modell
inception_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# InceptionV4-Modell importieren
inception_v4_model = timm.create_model('inception_v4', pretrained=True, num_classes=4)

# PyTorch-Optimierer und Verlustfunktion für InceptionV4
optimizer_inception_v4 = optim.Adam(inception_v4_model.parameters(), lr=0.001)
criterion_inception_v4 = nn.CrossEntropyLoss()

# Funktion zum Trainieren des Modells
def train_model(model, optimizer, criterion, train_loader, num_epochs):
    model.train()
    for epoch in range(num_epochs):
        total_loss = 0
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {total_loss / len(train_loader)}')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def test_model_inception_v4(model, test_loader):
    """
    Testet das InceptionV4-Modell auf den Testdaten.

    Parameter:
    model (torch.nn.Module): Das trainierte Modell.
    test_loader (torch.utils.data.DataLoader): Der DataLoader mit den Testdaten.

    Gibt zurück:
    test_accuracy (float): Die Testgenauigkeit des Modells.
    """
    # Modell in den Evaluierungsmodus setzen
    model.eval()
    
    # Variablen für die korrekten Vorhersagen und die Gesamtanzahl der Beispiele
    correct = 0
    total = 0
    
    # Berechnen Sie keine Gradienten während der Evaluierung
    with torch.no_grad():
        # Schleife über Testdaten
        for inputs, labels in test_loader:
            # Verschieben der Eingaben und Labels auf das richtige Gerät (GPU/CPU)
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            # Vorhersage mit dem Modell
            outputs = model(inputs)
            
            # Bestimmen Sie die vorhergesagten Klassen
            _, predicted = torch.max(outputs, 1)
            
            # Aktualisieren der Gesamtanzahl und der korrekten Vorhersagen
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    # Berechnen der Testgenauigkeit
    test_accuracy = correct / total
    
    return test_accuracy
```

```python
epochs = 10  # Anpassen Sie die Anzahl der Epochen an Ihre Anforderungen

train_images = []
train_labels = []

for _ in range(len(train_data) // batch_size):
    batch_images, batch_labels = train_data.next()
    train_images.append(batch_images)
    train_labels.append(np.argmax(batch_labels, axis=1))

train_images = np.vstack(train_images)
train_labels = np.hstack(train_labels)

# 10-fache Kreuzvalidierung
skf = StratifiedKFold(n_splits=10, shuffle=True)
fold = 1
accuracies_lenet5 = []
accuracies_alex = []
accuracies_inception = []
accuracies_inception_v4 = []

for train_index, val_index in skf.split(train_images, train_labels):
    train_fold_images, val_fold_images = train_images[train_index], train_images[val_index]
    train_fold_labels, val_fold_labels = train_labels[train_index], train_labels[val_index]

    # Trainingsdaten und Validierungsdaten erstellen
    train_fold_data = ImageDataGenerator().flow(train_fold_images, keras.utils.to_categorical(train_fold_labels),
                                                batch_size=batch_size)
    val_fold_data = ImageDataGenerator().flow(val_fold_images, keras.utils.to_categorical(val_fold_labels),
                                              batch_size=batch_size)

    history_lenet5 = lenet5.fit(train_fold_data, epochs=epochs, validation_data=val_fold_data, verbose= 0)
    acc_lenet5 = history_lenet5.history['val_accuracy'][-1]
    print(f'LeNet5 Fold: {fold}')
    accuracies_lenet5.append(acc_lenet5)
    
    history_alex = alexnet.fit(train_fold_data, epochs=epochs, validation_data=val_fold_data, verbose= 0)
    acc_alex = history_alex.history['val_accuracy'][-1]
    print(f'AlexNet Fold: {fold}')
    accuracies_alex.append(acc_alex)
    
    history_inception = inception_model.fit(train_fold_data, epochs=epochs, validation_data=val_fold_data, verbose=0)
    acc_inception = history_inception.history['val_accuracy'][-1]
    print(f'Inception-v3 Fold: {fold}')
    accuracies_inception.append(acc_inception)
    
    # Trainiere InceptionV4
    # Erstelle PyTorch DataLoader für Training und Validierung
    train_fold_loader = DataLoader(
        TensorDataset(torch.tensor(train_fold_images, dtype=torch.float32).permute(0, 3, 1, 2),
                      torch.tensor(train_fold_labels, dtype=torch.long)),
        batch_size=batch_size,
        shuffle=True
    )

    val_fold_loader = DataLoader(
        TensorDataset(torch.tensor(val_fold_images, dtype=torch.float32).permute(0, 3, 1, 2),
                      torch.tensor(val_fold_labels, dtype=torch.long)),
        batch_size=batch_size,
        shuffle=False
    )

    # Trainiere InceptionV4
    train_model(inception_v4_model, optimizer_inception_v4, criterion_inception_v4, train_fold_loader, epochs)
    # Teste InceptionV4 auf Validierungsdaten
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in val_fold_loader:
            outputs = inception_v4_model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    acc_inception_v4 = correct / total
    print(f'Inception-v4 Fold {fold} Validation Accuracy: {acc_inception_v4:.2f}')
    accuracies_inception_v4.append(acc_inception_v4)
    fold += 1

print(f'Mean Validation Accuracy LeNet5: {np.mean(accuracies_lenet5)}')
print(f'Mean Validation Accuracy AlexNet: {np.mean(accuracies_alex)}')
print(f'Mean Validation Accuracy Inception-v3: {np.mean(accuracies_inception)}')
print(f'Mean Validation Accuracy Inception-v4: {np.mean(accuracies_inception_v4)}')
```

```python
# Liste, um die PyTorch-Tensoren für Bilder und Labels zu sammeln
# Annahme: `test_data` ist ein Generator oder eine ähnliche Datenquelle für Testdaten
test_images = []
test_labels = []

# Iterieren Sie über die Testdaten, um Bilder und Labels zu sammeln
for _ in range(len(test_data) // batch_size):
    # Holen Sie sich eine Charge von Testbildern und Testlabels aus `test_data`
    batch_images, batch_labels = test_data.next()
    
    # Fügen Sie die Testbilder zu `test_images` hinzu
    test_images.append(batch_images)
    
    # Wandeln Sie die Labels in eindimensionale Arrays um, falls erforderlich, und fügen Sie sie zu `test_labels` hinzu
    # Hinweis: Der Vorgang, Labels in eindimensionale Arrays umzuwandeln, hängt von Ihrer Datenquelle ab.
    test_labels.append(np.argmax(batch_labels, axis=1))

# Konvertieren Sie die Listen `test_images` und `test_labels` in Arrays
test_images = np.vstack(test_images)
test_labels = np.hstack(test_labels)

test_dataset = TensorDataset(
    torch.tensor(test_images, dtype=torch.float32).permute(0, 3, 1, 2),
    torch.tensor(test_labels, dtype=torch.long)
)

# Erstellen eines DataLoaders für die Testdaten
# Hier verwenden wir shuffle=False, weil wir die Reihenfolge der Testdaten beibehalten wollen.
test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,  # Verwenden Sie die gleiche Batch-Größe wie beim Training
    shuffle=False  # Wir möchten die Reihenfolge der Testdaten beibehalten
)

# Evaluation auf Testdaten
test_loss_lenet5, test_acc_lenet5 = lenet5.evaluate(test_data)
print(f'Test Accuracy LeNet5: {test_acc_lenet5}')
          
test_loss_alex, test_acc_alex = alexnet.evaluate(test_data)
print(f'Test Accuracy AlexNet: {test_acc_alex}')

test_loss_inception, test_acc_inception = inception_model.evaluate(test_data)
print(f'Test Accuracy Inception-v3: {test_acc_inception}')

# Teste InceptionV4 auf Testdaten
test_accuracy = test_model_inception_v4(inception_v4_model, test_loader)
print(f'Test Accuracy Inception-v4: {test_accuracy}')
```

```python


# Setzen Sie die Anzahl der Klassen und Klassennamen auf Basis Ihrer Daten
num_classes = len(train_data.class_indices)
class_names = list(train_data.class_indices.keys())

def plot_confusion_matrix(conf_matrix, title, class_names):
    """ Plottet die Verwirrungsmatrix für ein Modell """
    plt.figure(figsize=(8, 8))
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45)
    plt.yticks(tick_marks, class_names)

    # Fügen Sie die Werte in die Verwirrungsmatrix ein
    thresh = conf_matrix.max() / 2.
    for i, j in itertools.product(range(conf_matrix.shape[0]), range(conf_matrix.shape[1])):
        plt.text(j, i, conf_matrix[i, j],
                 horizontalalignment="center",
                 color="white" if conf_matrix[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

# Berechnung und Visualisierung der Verwirrungsmatrizen für die Modelle

# Für LeNet5
test_predictions_lenet5 = np.argmax(lenet5.predict(test_images), axis=1)
conf_matrix_lenet5 = confusion_matrix(test_labels, test_predictions_lenet5)
plot_confusion_matrix(conf_matrix_lenet5, 'Confusion Matrix - LeNet5', class_names)

# Für AlexNet
test_predictions_alexnet = np.argmax(alexnet.predict(test_images), axis=1)
conf_matrix_alexnet = confusion_matrix(test_labels, test_predictions_alexnet)
plot_confusion_matrix(conf_matrix_alexnet, 'Confusion Matrix - AlexNet', class_names)

# Für InceptionV3
test_predictions_inception_v3 = np.argmax(inception_model.predict(test_images), axis=1)
conf_matrix_inception_v3 = confusion_matrix(test_labels, test_predictions_inception_v3)
plot_confusion_matrix(conf_matrix_inception_v3, 'Confusion Matrix - InceptionV3', class_names)

# Für InceptionV4
# Da InceptionV4 in PyTorch ist, müssen Sie einen anderen Ansatz wählen
inception_v4_model.eval()  # Setzen des Modells in den Evaluierungsmodus
test_predictions_inception_v4 = []

# Sammeln der Vorhersagen mit InceptionV4
with torch.no_grad():
    for inputs, _ in test_loader:
        # Verschieben der Eingaben auf das richtige Gerät (GPU/CPU)
        inputs = inputs.to(device)
        outputs = inception_v4_model(inputs)
        _, predicted = torch.max(outputs, 1)
        test_predictions_inception_v4.extend(predicted.cpu().numpy())

# Berechnung der Verwirrungsmatrix für InceptionV4
conf_matrix_inception_v4 = confusion_matrix(test_labels, test_predictions_inception_v4)
plot_confusion_matrix(conf_matrix_inception_v4, 'Confusion Matrix - InceptionV4', class_names)
```
